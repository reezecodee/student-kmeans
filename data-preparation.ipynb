```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('exam_test.csv')
```

```python
data.drop(['EthnicGroup', 'ParentEduc', 'TestPrep', 'LunchType', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild', 'NrSiblings', 'TransportMeans', 'WklyStudyHours'], axis=1, inplace=True)
data
```

```python
# Fitur yang digunakan
score_data = data[['Gender', 'MathScore', 'ReadingScore', 'WritingScore']]

# Hitung rata-rata nilai ujian
mean_scores = score_data.groupby('Gender').mean().reset_index().melt(id_vars='Gender', 
                                                                     var_name='Subject', 
                                                                     value_name='AverageScore')

# Visualisasi dengan barplot
plt.figure(figsize=(10, 6))
sns.barplot(data=mean_scores, x='Subject', y='AverageScore', hue='Gender')
plt.title('Rata-rata nilai ujian berdasarkan gender')
plt.xlabel('Mata pelajaran')
plt.ylabel('Nilai rata-rata')
plt.legend(title='Gender')
plt.show()
```

```python
# ubah data gender ke bentuk angka
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data['Gender'] = label_encoder.fit_transform(data['Gender'])  # female = 0, male = 1
data
```
